```python
import numpy as np
import matplotlib.pyplot as plt
import random as r
import sys as s


number_of_student = int(input("number of student"))
salary = float(input("salary"))
number_of_games = int(input("number of games"))
diff_per_round = float(input("budget"))

print(str(salary) + "  " + "最初的本金/賭資")

b = [salary]
print(b[0])

i=1
while i<=number_of_games:
    game_result=r.choice(np.arange(1,4))
    choose_of_participants=r.choices(np.arange(1,number_of_student+1),weights=None,cum_weights=None,k=2)
    if game_result==1 and (choose_of_participants[0]==1 or choose_of_participants[1]==1)  : #假設一人被選上且數字不變之機率為1/3＊2/c20取2
        b.append(b[-1])
    elif choose_of_participants[0]!=1 and choose_of_participants[1]!=1 :
        b.append(b[-1])
    elif game_result==2 and (choose_of_participants[0]==1 or choose_of_participants[1]==1) :
        b.append(b[-1]+diff_per_round)
    
    elif game_result==3 and (choose_of_participants[0]==1 or choose_of_participants[1]==1) :
        b.append(b[-1]-diff_per_round)
        
    i=i+1
    
    print(choose_of_participants)
    print(game_result)
    print(b)


    
    
    
plt.xlabel("student's budget")
plt.ylabel("number of students")
plt.hist(b,bins=100)
plt.show()



```
